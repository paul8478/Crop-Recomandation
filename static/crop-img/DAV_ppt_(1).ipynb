##**Step 6: Detecting and Filtering Outliers**

An outlier is a data point that significantly deviates from the rest of the dataset. It can be too high or too low compared to the majority of values.

**Effects of Outliers**

🔹 Skews Data Distribution – Outliers can distort the mean, making it unreliable.

🔹 Affects Model Performance – Many machine learning algorithms (especially linear regression, k-means clustering) are sensitive to outliers, leading to poor predictions.

🔹 Increases Variance – Outliers can make models more complex and less generalizable.

**How to Detect and Remove Outliers?**

There are two common techniques:

1️⃣ Z-Score (Standard Score Method)

2️⃣ Interquartile Range (IQR Method)

**1. Z-Score Method**

The Z-Score measures how far a data point is from the mean in terms of standard deviations.

**2. IQR (Interquartile Range) Method**

The IQR method works based on the spread of middle 50% of data.

Q1 (25th percentile) → First quartile

Q3 (75th percentile) → Third quartile

IQR = Q3 - Q1

📌 Rule of Thumb:

Outlier < Q1 - 1.5 × IQR

Outlier > Q3 + 1.5 × IQR

✅ Best for: Skewed data, non-Gaussian distributions.

❌ Not ideal for: Small datasets with few data points.



```python
from scipy.stats import zscore
# Select only numerical columns
num_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate Z-Scores
z_scores = np.abs(df[num_cols].apply(zscore))

# Find outliers (Z-Score > 3)
outliers_zscore = (z_scores > 3).sum()

print("\nOutliers detected using Z-Score:\n", outliers_zscore)
```

```python
# Visualize outliers before handling (Boxplot)
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[num_cols])
plt.title("Boxplot of Numerical Features (Before Removing Outliers)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Remove outliers using Z-Score threshold
df_zscore_cleaned = df[(z_scores < 3).all(axis=1)]

print("\nDataset Shape Before Removing Outliers (Z-Score):", df.shape)
print("Dataset Shape After Removing Outliers (Z-Score):", df_zscore_cleaned.shape)

# Visualize after removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_zscore_cleaned[num_cols])
plt.title("Boxplot of Numerical Features (After Removing Z-Score Outliers)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Detect outliers using IQR method
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] < lower_bound) | (df[column] > upper_bound)]

# Apply to numerical columns
outliers_iqr = {col: detect_outliers_iqr(df, col).shape[0] for col in num_cols}
print("\nOutliers detected using IQR:\n", outliers_iqr)

# Visualize outliers before removing using a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[num_cols])
plt.title("Boxplot of Numerical Features (Before Removing IQR Outliers)")
plt.xticks(rotation=45)
plt.show()
```

```python
# Function to remove IQR outliers
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Apply IQR filtering to all numerical columns
df_iqr_cleaned = df.copy()
for col in num_cols:
    df_iqr_cleaned = remove_outliers_iqr(df_iqr_cleaned, col)

print("\nDataset Shape Before Removing Outliers (IQR):", df.shape)
print("Dataset Shape After Removing Outliers (IQR):", df_iqr_cleaned.shape)

# Visualize after removing outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_iqr_cleaned[num_cols])
plt.title("Boxplot of Numerical Features (After Removing IQR Outliers)")
plt.xticks(rotation=45)
plt.show()
```

### **Total Energy Consumption Over Time (Top 5 Cities)**
